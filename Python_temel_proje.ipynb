* flatten

```python
ls1 = [[1,'a',['cat'],2],[[[3]],'dog'],4,[5,6],'7']
```

```python
def flatten(ls):
    '''
    boyutu farketmeksizin liste icindeki tum elemanlari duzlestirir.
    '''
    new_list =[]
    while len([True for e in ls if type(e)==list]) >0 :
        new_list =[]
        for item in ls:
            if type(item) == list:
                for i in item:
                    new_list.append(i)
                    
            else:
                new_list.append(item)
        
        ls = new_list
    return ls
```

```python
output = flatten(ls1)
output
```

* reverse

```python
ls2 = [[1, 2], [3, 4], [5, 6, 7]]
```

```python
def reverse(ls):
    '''
    ic ice 2 liste barindiran listelerde tersini yazdirabilir,
    derinlik 3 ve uzerine ciktiginda ic listeleri aynen yazdiracaktir.
    '''
    for i in ls:
        if type(i) == list:
            i.reverse()
    ls.reverse()
    return(ls)
  
```

```python
reverse(ls2)
```
